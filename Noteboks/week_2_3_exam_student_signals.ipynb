###**1. Creating and loading a dataset.**

```python
import pandas as pd

data = {
    "student_id" : ["2026CS001", "2026CS002", "2026IT001","2026IT002",
                  "2026CE001","2026CE002","2026CS003","2026IT003"],
    "name" : ["Aditi","Bhavana","Chaitra","Divya",
            "Esha","Fatima","Gauri","Harshada"],
    "department" : ["Computer Engineering","Computer Engineering",
                  "Information Technology","Information Technology",
                  "Civil Engineering","Civil Engineering",
                  "Computer Engineering","Information Technology"],
    "attendance_pct" : [92,64,78,55,88,71,96,43],
    "diary_entries" : [12,6,9,4,10,8,14,2],
    "tasks_completed" : [8,5,7,3,8,6,9,2],
    "assignment_score" : [84,58,72,46,79,66,91,35],
    "lab_score" : [88,61,76,52,82,69,94,41]
}

df = pd.DataFrame(data)

df.to_csv("week_2_3_exam_student_signals.csv", index=False)

df = pd.read_csv("week_2_3_exam_student_signals.csv")

df.head()
```

###Dataset Preview:
The dataset has been created and the first five records are displayed.

###**2. Data Inspection**

```python
print("\nData Types:\n", df.dtypes)

print("\nMissing values:\n", df.isnull().sum())

print("\nShape: ", df.shape)
```

### Data Inspection:
This code block checks the dataset for the data types, any missing values, dataset size.

### **3. Readiness Function**

```python
def readiness_score(attendance_pct, diary_enteries, tasks_completed, assignment_score, lab_score):
  score = attendance_pct * 0.30
  score += diary_enteries * 2
  score += tasks_completed * 3
  score += assignment_score * 0.20
  score += lab_score * 0.20

  return min(round(score, 2), 100)

df["readiness_score"] = df.apply(lambda row: readiness_score(row["attendance_pct"], row["diary_entries"], row["tasks_completed"], row["assignment_score"], row["lab_score"]), axis=1)

df[["name", "readiness_score"]]
```

### Readiness Score:
The readiness score is calculated by this function using attendace, diary entries, completed tasks, assignment scorea and lab scores from the dataset.

### **4. Classification**

```python
def performance_band(score):
    if score < 40:
        return "Support"

    elif score < 70:
        return "Developing"

    elif score < 90:
        return "Strong"

    else:
        return "Excellent"
bands = []

for score in df["readiness_score"]:
    bands.append(performance_band(score))

df["Performance_Band"] = bands

df[["name","readiness_score","Performance_Band"]]
```

###Performance Calculation:
Students are classified into Support, Developing, Strong, Excellent categories respectively.

### **5. Analytics**

```python
print("\nDepartment Average Readiness Score")

print(df.groupby("department")["readiness_score"].mean())

print("\nTop Performer")

top = df.loc[df["readiness_score"].idxmax()]
print(top[["name","department","readiness_score"]])

print("\nStudents Needing Support")

support = df[df["Performance_Band"]=="Support"]

print(support[["student_id","name","department","readiness_score"]])
```

###Final Analytics:
This notebook generates the department wise average readiness score of every student from the dataset, top performing student and students who need specical support.
